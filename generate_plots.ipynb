```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
import os
```

# Reinforce Plots

```python
# import os 
# import numpy as np
# import matplotlib.pyplot as plt
# from scipy.signal import savgol_filter

dir = 'reinforce'

results = [np.load(os.path.join(dir, file)) for file in os.listdir(dir)]
retults = results[:-1]
result_name = [file.split('.npy')[0] for file in os.listdir(dir)]

x_linspace = np.linspace(1, 2001, 2001)

for (result, result_name) in zip(results, result_name):
    learning_rate = result_name.split('_')[1]
    weight = result_name.split('_')[5]
    gamma = result_name.split('_')[3]
    result = savgol_filter(result, 101, 3)
    plt.figure()
    plt.plot(x_linspace, result.mean(axis=0),label=f'α={learning_rate}, w={weight}, γ={gamma}')
    plt.fill_between(x_linspace, result.mean(axis=0) - result.std(axis=0), result.mean(axis=0) + result.std(axis=0), alpha=0.2)
    plt.xlabel('Episode')
    plt.ylabel('Reward')
    plt.title('REINFORCE Experiments')
    plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
    plt.savefig(f'reinforce_plots/{result_name}.png', bbox_inches='tight')
```

```python
# import os 
# import numpy as np
# import matplotlib.pyplot as plt
# from scipy.signal import savgol_filter

dir = 'env_experiments'

results = [np.load(os.path.join(dir, file)) for file in os.listdir(dir)]
retults = results[:-1]
result_name = [file.split('.npy')[0] for file in os.listdir(dir)]

x_linspace = np.linspace(1, 2000, 2000)

for (result, result_name) in zip(results, result_name):
    rows = result_name.split('_')[1]
    cols = result_name.split('_')[3]
    speed = result_name.split('_')[5]
    obs_type = result_name.split('_')[8]
    
    result = savgol_filter(result, 101, 3)
    plt.figure()
    plt.plot(x_linspace, result.mean(axis=0),label=f'rows={rows}, cols={cols}, speed={speed}, obs_type={obs_type}')
    plt.fill_between(x_linspace, result.mean(axis=0) - result.std(axis=0), result.mean(axis=0) + result.std(axis=0), alpha=0.2)
    plt.xlabel('Episode')
    plt.ylabel('Reward')
    plt.title('Environment Experiments')
    plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
    plt.savefig(f'{dir}_plots/{result_name}.png', bbox_inches='tight')
```
